```python
Y = pd.get_dummies(Read.ReAdmis, drop_first=True)
Y.rename(columns = {'Yes':'ReAdmis'}, inplace = True)
Y.head()
```

```python
X = Read.drop('ReAdmis', axis=1)
X.head()
```

```python
X.info()
```
